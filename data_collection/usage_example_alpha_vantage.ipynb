```python
def get_stocksymbol(query):
    url = 'https://www.alphavantage.co/query?function=SYMBOL_SEARCH&keywords={}&apikey={}&datatype=json'.format(requests.utils.quote(query), av_config['api_key'])
    csv_op = requests.get(url)
    return pd.DataFrame(csv_op.json()['bestMatches'])
```
